```python
x = 1992
transf = []
while x < 2021: 
    path = 'C:\\Users\\Justyn\\CitadelDatathon\\data\\' + str(x)
    all_files = glob.glob(path + "/*.csv")
    li = []

    for filename in all_files:
        df = pd.read_csv(filename, index_col=None, header=0)
        li.append(df)

    frame = pd.concat(li, axis=0, ignore_index=True)
    transf.append(frame)
    x+=1

```

```python
transf[28]
```

```python
result = pd.concat(transf)
result
```
