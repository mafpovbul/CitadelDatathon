```python
import pandas as pd
import glob
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
```

```python
country = pd.read_csv("country.csv")
league = pd.read_csv("league.csv")
match = pd.read_csv("match.csv")
player = pd.read_csv("player.csv")
player_attributes = pd.read_csv("player_attributes.csv")
team_attributes = pd.read_csv("team_attributes.csv")
```

```python
x = 1992
transf = []
while x < 2021: 
    path = 'C:\\Users\\Administrator\\CitadelDatathon\\data\\' + str(x)
    all_files = glob.glob(path + "/*.csv")
    li = []

    for filename in all_files:
        df = pd.read_csv(filename, index_col=None, header=0)
        li.append(df)

    frame = pd.concat(li, axis=0, ignore_index=True)
    transf.append(frame)
    x+=1

```

```python
transf[28]
```

```python
result = pd.concat(transf)
result
```

```python
cleandf = result.drop_duplicates(subset = ['club_name','year'])
cleaned = cleandf.drop(columns = ['player_name','age','position','transfer_period','fee','season','transfer_movement','fee_cleaned','club_involved_name'])
cleaned
```

```python
clubnames = cleaned.club_name.unique()
```

```python
x=1992
while x<2021:
    y=0
    while y<len(clubnames):
        cleaned.loc[(cleaned['club_name'] == clubnames[y]) & (cleaned['year'] == x), 'fee_in'] = result.loc[(result['club_name'] == clubnames[y]) & (result['year'] == x) & (result['transfer_movement'] == 'in'), 'fee_cleaned'].sum() 
        y+=1
    x+=1
cleaned
```

```python
x=1992
while x<2021:
    y=0
    while y<len(clubnames):
        cleaned.loc[(cleaned['club_name'] == clubnames[y]) & (cleaned['year'] == x), 'fee_out'] = result.loc[(result['club_name'] == clubnames[y]) & (result['year'] == x) & (result['transfer_movement'] == 'out'), 'fee_cleaned'].sum() 
        y+=1
    x+=1
cleaned
```
