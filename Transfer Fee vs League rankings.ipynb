```python
match = pd.read_csv("match.csv")
league = pd.read_csv("league.csv")
team = pd.read_csv("team.csv", encoding = "latin-1")
match.league_id.unique()
league
```

```python
eplmatch = match[match["league_id"] == 1729]
eplmatch.season.unique()
```

```python
x=0
while x<len(match):
    if match.iloc[x]['home_team_goal'] == match.iloc[x]['away_team_goal']:
        match.at[x, 'home_team_points'] = 1
        match.at[x, 'away_team_points'] = 1
    elif match.iloc[x]['home_team_goal'] > match.iloc[x]['away_team_goal']:
        match.at[x, 'home_team_points'] = 3
        match.at[x, 'away_team_points'] = 0
    elif match.iloc[x]['home_team_goal'] < match.iloc[x]['away_team_goal']:
        match.at[x, 'home_team_points'] = 0
        match.at[x, 'away_team_points'] = 3
    x+=1
```

```python
yearlist = match.season.unique()
x = 0
yeardflist = []
while x < len(yearlist):
    year = match[match['season'] == yearlist[x]]
    home = year.groupby(["home_team_id"]).sum()
    away = year.groupby(["away_team_id"]).sum()
    y = 0
    while y<len(home):
        home.at[home.index[y], 'total_points'] = home.at[home.index[y], "home_team_points"] + away.at[away.index[y], "away_team_points"]
        y+=1
    z=0
    while z<len(home):
        home.at[home.index[z], 'team_id'] = home.index[z]
        z+=1
    merged = pd.merge(team,home, on = 'team_id')
    merged1 = merged[["team_id","total_points"]]
    yeardflist.append(merged1)
    x+=1
```

```python
a = yeardflist[0]
a[a["team_id"] == 8634]
```

```python
teamtoleague = match[["league_id","home_team_id"]]
ttl = teamtoleague.drop_duplicates(subset = ['home_team_id'])
tttl = ttl.rename(columns={"home_team_id" : "team_id"})
```

```python
x = 0
while x<len(yearlist):
    a = yeardflist[x]
    a["season"] = yearlist[x]
    x+=1
teamfull = pd.concat(yeardflist, ignore_index = True)
tf = pd.merge(teamfull,tttl, on = "team_id")
tf
```

```python
ttf = pd.merge(tf,team, on = "team_id")
ttf
```

```python
x = 0
while x <len(ttf):
    ttf.loc[x, "year"] = int(ttf.loc[x, "season"][0:4])
    x+=1
tttf = ttf.drop(columns = ['season'])
tttf
```

```python
df = df.reset_index(drop=True)
x=0
while x<len(df):
    df.loc[x, "net_fee"] = df.loc[x, "fee_out"] - df.loc[x, "fee_in"]
    x+=1
df
```

```python
ddf = df[df["year"]>=2008]
abcd = pd.read_excel("clubnames.xlsx")
xyz = abcd.columns.values.tolist()
abcde = abcd.drop(columns = xyz[0])
ab = pd.merge(ddf, abcde, on = "club_name")
ab
```

```python
yes = ab.merge(tttf, how='inner', left_on=["club_name", "year"], right_on=["team_long_name","year"])
yes.league_name.unique()
```

```python
epl = yes[yes["league_name"] == "Premier League"]
epl
```

```python
yearlist= epl.year.unique()
x = 0
dflist = []

while x<len(yearlist):
    eight = epl[epl["year"] == yearlist[x]]
    eight.sort_values(by = ["total_points"], ascending = False)
    eight1 = eight.reset_index(drop = True)
    y = 1
    while y<len(eight1):
        eight1.loc[y, "ranking"] = y
        y+=1
    dflist.append(eight1)
    x+=1
for x in dflist:
    plt.figure()
    sb.barplot(data = x, y="ranking", x="net_fee", orient = "h")
```
