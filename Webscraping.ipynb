```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq
import requests
```

```python
url = 'https://www.premierleague.com/tables?co=1&se=274&ha=-1'
page = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'})

#html parsing
page_soup = soup(page.content, "html.parser")
page_soup
```

```python
names_containers = page_soup.find_all('span', class_ = "value")
names_containers
```

```python
names_containers = page_soup.find_all('td', class_ = "team")
names_containers[1]
```

```python
pos_containers = page_soup.find_all('td', class_ = "pos-transfer-cell")
print(len(pos_containers))
pos_containers[0].text
```

```python
value_containers = page_soup.find_all('td', class_ = "rechts mw-transfer-cell")
print(len(value_containers))
value_containers[0].text
if value_containers[1].text[-1] == '.':
    print("x")
```

```python
club_containers = page_soup.find_all('td', class_ = "no-border-links verein-flagge-transfer-cell")
print(len(club_containers))
club_containers[0].a.text
```
